```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pyzbar.pyzbar import decode
from pprint import pprint
```

```python
image = cv2.imread('img/qr5.png', cv2.IMREAD_COLOR_RGB)

qr_info = decode(image)

pprint(qr_info)
```

```python
image = cv2.imread('img/qr5.png', cv2.IMREAD_COLOR_RGB)

qr_info = decode(image)

print(qr_info)

for qr in qr_info:
    print(qr.data)
    cv2.rectangle(image, (qr.rect.left - 5, qr.rect.top -5), (qr.rect.left + qr.rect.width + 5, qr.rect.top + qr.rect.height + 5), (0, 255, 0), 3)
    cv2.polylines(image, [np.array(qr.polygon)], True, (0, 0, 255), 3)
plt.imshow(image)
```
